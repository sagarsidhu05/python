***THEORY QUESTIONS***

1.What is a parameter?
ANS=A parameter is a model component that is learned from data during training. It defines the behavior of the model, such as weights in linear regression or neural networks.

2.What is correlation?
ANS=Correlation measures the linear relationship between two variables. It ranges from -1 to 1. A positive value means both increase together; negative means one increases as the other decreases.

3.What does negative correlation mean?
ANS=Negative correlation means that as one variable increases, the other tends to decrease. For example, increased exercise is often associated with lower body weight.

4.Define Machine Learning. What are the main components in Machine Learning?
ANS=Machine Learning is a method where systems learn from data to make predictions or decisions. Key components are data, features, model, loss function, optimizer, and evaluation metrics.

5.How does loss value help in determining whether the model is good or not?
ANS=Loss measures the error between predicted and actual values. A lower loss indicates a better-performing model. It guides the model during training to improve accuracy.

6.What are continuous and categorical variables?
ANS=Continuous variables are numerical and can take infinite values (e.g., height). Categorical variables represent distinct groups or categories (e.g., gender, color).

7.How do we handle categorical variables in Machine Learning? What are the common techniques?
ANS=Categorical variables are converted to numbers using encoding techniques like Label Encoding, One-Hot Encoding, or Ordinal Encoding. This makes them usable by ML models.

8.What do you mean by training and testing a dataset?
ANS=The dataset is split into training and testing sets. The model learns from the training set and is evaluated on the test set to check generalization.

9.What is sklearn.preprocessing?
ANS=This is a module in scikit-learn for preprocessing tasks like scaling, encoding, and normalization. It prepares data for efficient and accurate model training.

10.What is a Test set?
ANS=The test set is a portion of data not seen by the model during training. It is used to evaluate the model’s real-world performance and generalization.

11.How do we split data for model fitting (training and testing) in Python?
ANS=In Python, data is split using train_test_split() from sklearn.model_selection. It separates data into training and test sets, typically using an 80:20 ratio.

12.How do you approach a Machine Learning problem?
ANS=Understand the problem, gather data, clean and explore it, preprocess features, split data, train models, evaluate them, and deploy the best one.

13.Why do we have to perform EDA before fitting a model to the data?
ANS=EDA helps understand data patterns, spot outliers, and guide preprocessing. It improves model performance and prevents errors by revealing data issues early.

14.What is correlation?
ANS=Correlation shows how strongly two variables are related. It is useful in identifying relationships between features and target variables in data.

15.What does negative correlation mean?
ANS=It means one variable decreases when the other increases. It helps detect inverse relationships between features.

16.How can you find correlation between variables in Python?
ANS=Use .corr() on a DataFrame to compute correlation. Use heatmaps (seaborn.heatmap) for visual representation of relationships between variables.

17.What is causation? Explain difference between correlation and causatio1n with an example.
ANS=Causation means one variable directly affects another, while correlation only shows a relationship. Example: More AC usage correlates with ice cream sales, but doesn’t cause them.

18.What is an Optimizer? What are different types of optimizers? Explain each with an example.
ANS=Optimizers adjust model parameters to reduce loss. Common types include SGD, Adam, RMSprop, and Adagrad. Each has its own learning rate and update strategy.

19.What is sklearn.linear_model ?
ANS=A scikit-learn module with linear models like LinearRegression and LogisticRegression. It provides tools for regression and classification problems.

20.What does model.fit() do? What arguments must be given?
ANS=Fits the model to training data. It takes input features (X_train) and target (y_train) to learn the relationship between them.

21.What does model.predict() do? What arguments must be given?
ANS=Uses the trained model to make predictions on new or test data. It returns estimated outputs based on learned patterns.

22.What are continuous and categorical variables?
ANS=Continuous variables are numerical values that can take any value within a range, including decimals (e.g., height, temperature, salary).
Categorical variables represent discrete groups or categories without numeric meaning (e.g., gender, color, country).
Continuous variables are used in regression problems, while categorical variables are often used in classification tasks.
Categorical data must be encoded before being used in most machine learning algorithms.

23.What is feature scaling? How does it help in Machine Learning?
ANS=Feature scaling normalizes data so features contribute equally. It improves model performance, especially in distance-based or gradient-based algorithms.

24.How do we perform scaling in Python?
ANS=Use StandardScaler or MinMaxScaler from sklearn.preprocessing to scale features. It standardizes features to a fixed range or mean/variance.

25.What is sklearn.preprocessing?
ANS=Used for encoding, scaling, imputing, and transforming data into formats suitable for model training.

26.How do we split data for model fitting (training and testing) in Python?
ANS=Handled using train_test_split() to divide data into training and testing sets for model training and validation.

27.Explain data encoding?
ANS=Transforms categorical data into numeric format using techniques like Label Encoding and One-Hot Encoding, making it suitable for machine learning models.
