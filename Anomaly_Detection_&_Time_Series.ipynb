#**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

- **Anomaly Detection** is the process of identifying rare items, events, or observations which differ significantly from the majority of the data. These anomalies often indicate critical incidents such as fraud, system failures, or security breaches.

It is widely used in areas like:
- Credit card fraud detection
- Medical diagnosis
- Network intrusion detection
- Machine fault prediction


##  Types of Anomalies

### 1️. Point Anomaly (Global Outlier)

A **single data point** that deviates significantly from the rest of the dataset.

- **Example**: In a dataset of daily credit card transactions where the average value is ₹500, a transaction of ₹10,000 would be a point anomaly.


### 2️.Contextual Anomaly (Conditional Outlier)

A data point is considered anomalous **only in a specific context**, such as time or location.

- **Example**: A temperature of 30°C is normal in **May** but would be a contextual anomaly in **December**.


### 3️. Collective Anomaly (Group Outlier)

A **group of related data points** is anomalous even if individual data points are not.

- **Example**: A sudden spike in network activity from multiple IP addresses at once (like in a DDoS attack) is a collective anomaly.

---

##  Summary Table

| Type               | Description                                           | Example                                                  |
|--------------------|-------------------------------------------------------|-----------------------------------------------------------|
| **Point Anomaly**      | Single outlier in data                              | ₹10,000 transaction in a ₹500 dataset                     |
| **Contextual Anomaly** | Outlier in a specific context (time, location)      | 30°C in winter                                            |
| **Collective Anomaly** | Group of values that are anomalous together         | Group of IPs flooding a server simultaneously             |

# **Question 3: What are the key components of a Time Series? Explain each with one example.**
- A time series is a sequence of data points collected or recorded at specific time intervals such as daily, weekly, monthly, or yearly. It is commonly used in fields like finance, weather forecasting, stock market analysis, and sensor data monitoring.

A time series typically consists of the following four key components:

1. Trend
The trend is the long-term movement or direction in the data over time. It can be upward, downward, or flat depending on the overall direction of the data.

- Example: The steady increase in global average temperatures over the past several decades represents an upward trend.

2. Seasonality
Seasonality refers to regular and predictable patterns that repeat over a fixed period, such as daily, weekly, or annually. These are usually caused by external factors like weather, holidays, or social habits.

- Example: Sales of air conditioners increasing every summer is a clear example of seasonality.

3. Cyclic Component
The cyclic component includes fluctuations in data that are not of a fixed period like seasonality. These cycles often occur over long durations (more than a year) and are influenced by economic or political conditions.

- Example: Economic booms and recessions that affect stock market performance show cyclical patterns.

4. Irregular Component (Noise)
The irregular component, also known as residual or noise, represents random, unpredictable variations in the data. These variations do not follow a pattern and cannot be explained by trend, seasonality, or cyclic effects.

- Example: A sudden drop in electricity consumption due to a natural disaster like an earthquake is considered irregular.

#**Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**
## What is Stationarity?

- A **stationary time series** is one whose **statistical properties** such as **mean**, **variance**, and **autocorrelation** are **constant over time**.In simple terms, the data values fluctuate around a constant level and have a constant spread (variance), making them predictable over time.


- ### Why is Stationarity Important?
Most time series models, like **ARIMA**, assume that the series is stationary. If not, the model might not work correctly or produce misleading forecasts.

## How to Test for Stationarity?

1. **Visual Inspection**:
   - Plot the time series.
   - Trends, seasonality, or changing variance suggest non-stationarity.

2. **Summary Statistics**:
   - Split the series into parts and compare the mean and variance of each part.

3. **Statistical Tests**:

   - **Augmented Dickey-Fuller (ADF) Test**  
     - **Null Hypothesis (H₀)**: The series is **non-stationary**  
     - **If p-value < 0.05**, reject H₀ → the series is **stationary**

   - **KPSS Test**  
     - **Null Hypothesis (H₀)**: The series is **stationary**  
     - **If p-value < 0.05**, reject H₀ → the series is **non-stationary**


## How to Transform a Non-Stationary Series to Stationary

1. **Differencing**  
   - Subtract current value from previous value:  
     `Y_t' = Y_t - Y_(t-1)`
   - First-order differencing usually removes trend.

2. **Log Transformation**  
   - Helps stabilize variance:  
     `log(Y_t)`

3. **Detrending**  
   - Remove the trend component using linear regression or subtracting a moving average.

4. **Seasonal Differencing**  
   - Subtract the value from the same season in the previous cycle:  
     `Y_t - Y_(t-12)` for monthly data.


##Summary Table

| **Method**              | **Purpose**                          | **Example**                  |
|--------------------------|--------------------------------------|-------------------------------|
| Differencing             | Remove trend                        | `Y_t - Y_(t-1)`              |
| Log Transformation       | Stabilize variance                  | `log(Y_t)`                   |
| Detrending               | Remove linear/nonlinear trend       | Subtract moving average      |
| Seasonal Differencing    | Remove repeating seasonal patterns  | `Y_t - Y_(t-12)`             |

# **Question5:Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**
The following table summarizes the differences among AR, MA, ARIMA, SARIMA, and SARIMAX models based on structure and application:

| **Model**   | **Full Form**                             | **Structure / Components**                                      | **Suitable For**                                         | **Example Use Case**                             |
|-------------|-------------------------------------------|------------------------------------------------------------------|-----------------------------------------------------------|---------------------------------------------------|
| **AR**      | Autoregressive                            | Uses past values (lags): AR(p)                                  | Stationary series with autocorrelation                    | Forecasting temperature from past values          |
| **MA**      | Moving Average                            | Uses past errors (residuals): MA(q)                             | Stationary series with random noise                       | Modeling short-term shocks in economic data       |
| **ARIMA**   | Autoregressive Integrated Moving Average  | Combines AR + I (differencing) + MA: ARIMA(p, d, q)             | Non-stationary series without seasonality                 | Predicting monthly sales                          |
| **SARIMA**  | Seasonal ARIMA                            | ARIMA + seasonal components: ARIMA(p, d, q)(P, D, Q, s)         | Non-stationary series with seasonality                    | Forecasting monthly electricity demand            |
| **SARIMAX** | Seasonal ARIMA with Exogenous Variables   | SARIMA + external variables (exog)                              | Seasonal series influenced by external features           | Forecasting sales influenced by marketing budget  |



### Notes:

- **p, d, q**: AR, differencing, and MA orders
- **P, D, Q, s**: Seasonal orders and seasonal period
- **Exogenous variables** (in SARIMAX): Variables outside the time series (e.g., weather, ads)

#**Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import warnings
warnings.filterwarnings('ignore')


df=pd.read_csv('NYC_taxi_fare_data.csv')
df.head()



```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('NYC_taxi_fare_data.csv')

numeric_df = df.select_dtypes(include='number').dropna()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

model = IsolationForest(n_estimators=100, contamination=0.02, random_state=42)
pred = model.fit_predict(scaled_data)

numeric_df['anomaly'] = pred

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
numeric_df['pca1'] = pca_result[:, 0]
numeric_df['pca2'] = pca_result[:, 1]

plt.figure(figsize=(10, 6))
plt.scatter(numeric_df[numeric_df['anomaly'] == 1]['pca1'],numeric_df[numeric_df['anomaly'] == 1]['pca2'],c='blue', label='Normal', alpha=0.6)
plt.scatter(numeric_df[numeric_df['anomaly'] == -1]['pca1'],numeric_df[numeric_df['anomaly'] == -1]['pca2'],c='red', label='Anomaly', alpha=0.8)
plt.title('Isolation Forest Anomaly Detection (PCA 2D Projection)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.grid(True)
plt.show()
```

#**Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.**

```python

import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

df= pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month')
model = SARIMAX(df['#Passengers'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit(disp=False)

forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

forecast_index = pd.date_range(start='1961-01-01', periods=12, freq='MS')

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['#Passengers'], label='Historical Data')
plt.plot(forecast_index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='red', alpha=0.2, label='95% Confidence Interval')
plt.title('Airline Passengers Forecast (SARIMA)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

# **Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import LocalOutlierFactor

iris = load_iris()
X = iris.data[:, [0, 2]]

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

outliers = y_pred == -1
inliers = y_pred == 1

plt.figure(figsize=(8, 6))
plt.scatter(X[inliers, 0], X[inliers, 1], c='blue', label='Inliers')
plt.scatter(X[outliers, 0], X[outliers, 1], c='red', label='Outliers', marker='x')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('LOF Anomaly Detection on Iris Dataset')
plt.legend()
plt.show()
```

###  1. Anomaly Detection in Streaming Data

**Recommended Algorithm: Isolation Forest**

- Efficient for real-time and high-dimensional data.
- Detects both **spikes** and **drops** in consumption.
- Can be retrained incrementally and quickly.

**Why not LOF or DBSCAN?**

- `LOF` is slower and less scalable for streaming.
- `DBSCAN` is non-incremental and not ideal for real-time use.

-**Workflow:**
- Stream data (energy usage, timestamp, weather).
- Standardize features with `StandardScaler`.
- Apply trained `IsolationForest` model to assign anomaly scores.
- Flag records as anomalies when score < threshold.
- Trigger alerting system (email, dashboard, auto-scaling, etc.).

---

### ⏱2. Short-Term Forecasting Model

**Recommended Model: SARIMAX**

- Captures **seasonality** (daily, weekly patterns).
- Includes **external variables** like temperature, humidity, and region.
- More flexible than SARIMA and ARIMA.

**Why not ARIMA/SARIMA?**

- `ARIMA`: doesn't support seasonality or external variables.
- `SARIMA`: seasonal, but doesn’t handle external regressors (weather, region).

**Forecasting Setup:**
- Use `energy_usage` as target `y`.
- Add `temperature`, `region`, `hour_of_day`, etc. as exogenous `X`.
- Predict 1–2 hours ahead (4–8 steps of 15 min intervals).
---

### 3. Validation & Monitoring

**Offline Validation:**
- Forecast: Use **MAE**, **RMSE**.
- Anomaly detection: **Precision**, **Recall**, **F1-score**.

**Real-Time Monitoring:**
- Monitor rolling RMSE/MAE.
- Alert if:
  - Prediction error exceeds threshold.
  - Anomaly rate rises sharply.
- Periodically retrain models (weekly/monthly) using new streaming data.

**Tools:**
- Use **MLflow** for experiment tracking.
- Use **Prometheus + Grafana** for dashboards and alerts.

---

### 4. Business Impact

**Benefits:**

- Anomaly detection helps catch faults or energy theft early.
- Enables **dynamic pricing** during peak demand.
- Improves **grid efficiency**, saving costs and energy.
- Supports **regulatory compliance** by ensuring supply stability.
