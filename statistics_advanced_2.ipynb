***THEORY QUESTION***

1.What is hypothesis testing in statistics?

ANS=Hypothesis testing is a method used in statistics to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (hypothesis) to determine whether there is enough evidence to reject it.

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

ANS=Null Hypothesis (H₀): The default or no-effect assumption (e.g., no difference, no change).
Alternative Hypothesis (H₁ or Ha): The statement you want to test for; it contradicts the null (e.g., there is a difference or effect).

3.What is the significance level in hypothesis testing, and why is it important?

ANS=The significance level (α) is the threshold for rejecting the null hypothesis, commonly set at 0.05 (5%). It represents the probability of making a Type I error (rejecting a true null hypothesis).

4.What does a P-value represent in hypothesis testing?

ANS=The P-value is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from your sample data, assuming the null hypothesis is true.

5.How do you interpret the P-value in hypothesis testing?

ANS=If P-value ≤ α, reject the null hypothesis (evidence supports the alternative).

If P-value > α, do not reject the null hypothesis (not enough evidence).

6.What are Type 1 and Type 2 errors in hypothesis testing?
ANS=Type I Error (α): Rejecting a true null hypothesis.

Type II Error (β): Failing to reject a false null hypothesis.

7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

ANS=One-tailed test: Tests for an effect in one direction only.

Two-tailed test: Tests for an effect in both directions (more conservative).

8.What is the Z-test, and when is it used in hypothesis testing?

ANS=A Z-test is used when:
The population standard deviation (σ) is known.
The sample size is large (n > 30).
It tests hypotheses about population means or proportions.

9.How do you calculate the Z-score, and what does it represent in hypothesis testing?

ANS=Z = (X̄ - μ) / (σ / √n)
It measures how many standard deviations the sample mean (X̄) is from the population mean (μ).

10.What is the T-distribution, and when should it be used instead of the normal distribution?

ANS=The T-distribution is used when:
The sample size is small (n < 30).
The population standard deviation is unknown.
It accounts for extra uncertainty in small samples.

11.What is the difference between a Z-test and a T-test?
ANS=Z-test: Known σ, large sample.
T-test: Unknown σ, small sample.
Both compare sample means to population means but use different distributions.

12.What is the T-test, and how is it used in hypothesis testing?
ANS=A T-test compares means:
One-sample T-test: Sample mean vs. known value.
Two-sample T-test: Compares two sample means.
Paired T-test: Compares means from the same group at different times.

13.What is the relationship between Z-test and T-test in hypothesis testing?

ANS=Both test means and use similar formulas, but the Z-test uses the normal distribution while the T-test uses the T-distribution (more spread out for small samples).

14.What is a confidence interval, and how is it used to interpret statistical results?

ANS=A confidence interval (CI) gives a range of values within which the population parameter is likely to lie. For example, a 95% CI means you’re 95% confident the true mean lies in that range.

15.What is the margin of error, and how does it affect the confidence interval?

ANS=The margin of error (MOE) is half the width of the confidence interval. A larger MOE means less precise estimates, influenced by sample size and variability.

16.How is Bayes' Theorem used in statistics, and what is its significance?

ANS=Bayes' Theorem updates the probability of a hypothesis based on new evidence.
It is key in Bayesian statistics, allowing prior knowledge to be combined with data.

17.What is the Chi-square distribution, and when is it used?

ANS=The Chi-square (χ²) distribution is used to test relationships between categorical variables and assess how well observed data fit expected distributions.

18.What is the Chi-square goodness of fit test, and how is it applied?

ANS=It tests whether the observed frequency distribution of a categorical variable matches an expected distribution. Formula:
χ² = Σ[(O - E)² / E], where O = observed, E = expected.

19.What is the F-distribution, and when is it used in hypothesis testing?

ANS=The F-distribution is used to compare variances. It is the basis for:
ANOVA (analysis of variance)
F-tests for comparing models

20.What is an ANOVA test, and what are its assumptions?

ANS=ANOVA tests whether the means of 3 or more groups are significantly different.
Assumptions:
Normality
Homogeneity of variance
Independence of observations

21.What are the different types of ANOVA tests?

ANS=One-way ANOVA – One independent variable.
Two-way ANOVA – Two independent variables.
Repeated measures ANOVA – Same subjects measured multiple times.

22.What is the F-test, and how does it relate to hypothesis testing?

ANS=The F-test compares two variances or model fits. In ANOVA, it tests if at least one group mean is significantly different. Large F-values suggest rejecting the null hypothesis.

1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import scipy.stats as stats
import numpy as np

# Sample data
sample = [5.1, 5.5, 5.3, 5.8, 5.0, 5.4]
pop_mean = 5.0
pop_std = 0.4  # known population standard deviation
n = len(sample)

# Z-test
sample_mean = np.mean(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
import numpy as np
from scipy import stats

# Step 1: Define sample data and known population parameters
sample_data = [101, 98, 100, 97, 103, 99, 102, 98, 100, 96]  # Sample data
pop_mean = 100     # Known population mean (μ)
pop_std = 3        # Known population standard deviation (σ)
n = len(sample_data)

# Step 2: Calculate sample mean
sample_mean = np.mean(sample_data)

# Step 3: Calculate the Z-score
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# Step 4: Calculate the two-tailed P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 5: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Interpret result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
def plot_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    se = sigma / np.sqrt(n)
    z_alpha = stats.norm.ppf(1 - alpha)

    x = np.linspace(-4, 4, 1000)
    y0 = stats.norm.pdf(x, mu0, se)
    y1 = stats.norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0 Distribution', color='blue')
    plt.plot(x, y1, label='H1 Distribution', color='orange')

    x_crit = mu0 + z_alpha * se
    plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')

    plt.fill_between(x, y0, where=(x > x_crit), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, y1, where=(x < x_crit), color='green', alpha=0.3, label='Type II Error (β)')

    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.show()

plot_type_errors()
```

9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return mean - margin, mean + margin

data = np.random.normal(75, 8, 30)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

10.Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)
    z_crit = stats.norm.ppf((1 + confidence) / 2)
    return z_crit * std_err

# Example usage
sample = np.random.normal(50, 5, 30)
moe = margin_of_error(sample)
print(f"Margin of Error: {moe:.3f}")
```

11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B

# Example: Disease testing
# P(Disease) = 0.01, P(Pos | Disease) = 0.99, P(Pos) = 0.05
posterior = bayes_theorem(0.01, 0.99, 0.05)
print(f"Posterior probability (P(Disease | Positive test)): {posterior:.4f}")
```

22.Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=55, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)

# Step 2: Perform one-way ANOVA
f_stat, p_val = stats.f_oneway(group1, group2, group3)

# Step 3: Print ANOVA results
print(f"One-Way ANOVA Results")
print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

# Step 4: Visualize with boxplots
plt.figure(figsize=(8, 6))
plt.boxplot([group1, group2, group3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Comparison of Group Means (One-Way ANOVA)')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
# Group 1: 45/100, Group 2: 30/80
p1, n1 = 0.45, 100
p2, n2 = 0.375, 80

p_combined = (p1 * n1 + p2 * n2) / (n1 + n2)
se = np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
z = (p1 - p2) / se
p_val = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-stat: {z:.3f}, P-value: {p_val:.4f}")
```
